Question no 3.Integer data type:

      An integer data type in Python is used to store whole numbers (numbers without decimal points).
     It can be positive, negative, or zero.

Example

marks = 85

print(marks)


Question no 4.float data type:

     A float data type in Python is used to store decimal (fractional) numbers.
    It contains numbers with a decimal point.

    example:
    
    height = 5.6
    
     print(height)


Question no 5.String datatype:

      A string data type in Python is used to store text or characters.
      Strings are written inside quotation marks.

      Two Ways to Create a String in Python

 1. Using Single Quotes ' '

  name = 'Amina'
  
  print(name)


 2. Using Double Quotes " "

  city = "Lahore"

  print(city)

Question 7.()Function:

     The type() function in Python is used to check the data type of a variable or value.
     It tells you whether a variable is an integer, float, string, boolean, etc

Example:

Syntax:

type(variable_name)

Example:

 age = 20

 print(type(age))

Question 8.input()function

     The input() function is used to take input from the user while the program is running.

Key Points:

 1) input() pauses the program and waits for the user to type something.

2) The default datatype of the value taken by input() is always string (str), even if the user enter      a   number.

Example:
 
name = input("Enter your name: ")

print("Hello,", name)

print(type(name))


Question 9.type casting:

      Type casting is the process of converting one data type into another in Python.

      For example: converting a string to an integer, or a float to a string, etc.

  Python Code to Convert a String into an Integer

    num_str = "25"         
    num_int = int(num_str) 

    print(num_int)         
    print(type(num_int))
