## EXAMPLES


```python
string = "aabbcc"
count = 3

while True:
    if string[0] == 'a':
        string = string[2:]
    elif string[-1] == 'b':
        string = string[:2]
    else:
        count += 1
        break

print( string)
print( count)
```

This program uses a loop to check and modify a string. If the string starts with the character 'a', the first two characters are removed from the string. If the condition is not satisfied, the program increases the value of the count and stops the loop. At the end, the program prints the final string and the updated count value.

## Break statement

```python
i=1
while True:
    print(i)
    if i==5:
       break
    i +=1    
```

the loop runs continuously and increases the value of i by 1 in each iteration when i beacomes 5 the break statement stops the loop imediatly

## Continue statement

```python
i=0
while i<5:
    i+=1
    if i==3:
        continue
    print(i)   
```

continue statement skips that step when condition is true it goes back to start of the loop and continue next turn.
