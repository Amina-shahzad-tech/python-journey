## Match statement
        
        match statement is a descion making statement whose work is to check the values and make descion according to these values

# program using match statement


```python
fruit=input("enter a fruit name")
match fruit:
    case "apple":
        print("apple")
    case "orange":
        print("orange")
    case "mango":
        print("mango")
    case _: 
        print ("not known fruit")        
```

```python
num=input("enter number 1,2,3")
match num:
    case 1:
        print("1")
    case 2:
        print("2")
    case 3:
        print("3")
    case _:
        print("wrong number")            
```

Question:

differnce between match and if else statement

## match statement

1) used for matching values or patterens.
2) mostly checks specific values.
3) best when comparing one variable to many.
4) uses match and case.

## if else statement

1) used for conditions.
2) can check any condition(>,<,== logic etc.).
3) very flexible.
4) uses if,elif,else.



```python
marks = 78

if marks >= 50:
    if marks >= 80:
        print("Grade A")
    else:
        print("Grade B")
else:
    print("Fail")
```

```python
age = 16

if age >= 13:
    if age >= 18:
        print("Adult")
    else:
        print("Teenager")
else:
    print("Child")
```

```python
num = -5

if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive Number")
else:
    print("Negative Number")
```

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login Successful")
    else:
        print("Wrong Password")
else:
    print("Wrong Username")
```

```python
number = 12

if number % 2 == 0:
    if number > 10:
        print("Even and Greater than 10")
    else:
        print("Even but Less than or Equal to 10")
else:
    print("Odd Number")
```
